### 5. (Função com retorno com parâmetro) Faça um programa contendo uma função/método que receba um valor de porcentagem de aumento e um salário via parâmetro, aplique este aumento ao salário do funcionário.
###Na parte principal (main) do programa, utilize um laço de repetição para ler 10 salários, chame a função para aplicar o aumento e gerar o novo salário, ainda dentro da estrutura de repetição acumule os novos salários e ao final apresente quanto será gasto no novo salário.
###Também apresente qual será a diferença entre o que se pagava para todos os funcionário e o que será pago após o aumento.

```python
# Digite aqui seu código
def aplicar_aumento(salario, percentual_aumento):
    """Aplica o aumento ao salário e retorna o novo salário."""
    return salario + (salario * (percentual_aumento / 100))

# Inicialização das variáveis de controle
total_salarios_anteriores = 0
total_salarios_novos = 0

# Entrada do percentual de aumento
percentual = float(input("Digite o percentual de aumento (%): "))

# Laço para processar 10 salários
for i in range(1, 11):
    salario_atual = float(input(f"Digite o salário do {i}º funcionário: "))

    novo_salario = aplicar_aumento(salario_atual, percentual)  # Chamada da função

    total_salarios_anteriores += salario_atual
    total_salarios_novos += novo_salario

# Cálculo da diferença total
diferenca = total_salarios_novos - total_salarios_anteriores

# Exibição dos resultados
print(f"\nTotal gasto com salários antes do aumento: R$ {total_salarios_anteriores:.2f}")
print(f"Total gasto com salários após o aumento: R$ {total_salarios_novos:.2f}")
print(f"Impacto financeiro do aumento: R$ {diferenca:.2f}")
```

### 6. (Função com retorno com parâmetro) Faça um programa contendo algumas funções:
###a) Crie uma função para apresentar o menu e retornar a opção/número/cálculo escolhido.

    Menu de cálculos
    1.   Número ao quadrado
    2.   Número ao cubo
    3.   Raiz do número
    4.   Raiz cúbica do número
    Qual é a opção desejada?

###b) Desenvolva uma função para cada opção de cálculo.

###c) A função main() chamará todas as outras.

###Observação: Na chamada das funções deve-se utilizar uma estrutura de repetição, que permita diversos cálculos, quando o usuário quiser sair da calculadora digitará qualquer valor diferente dos **números** do menu.


```python
def apresentar_menu():
    """Apresenta o menu e retorna a opção escolhida pelo usuário."""
    print("\nMenu de cálculos:")
    print("1. Número ao quadrado")
    print("2. Número ao cubo")
    print("3. Raiz do número")
    print("4. Raiz cúbica do número")
    print("Para sair digite qualquer outro numero")
    return int(input("Qual é a opção desejada? "))

def ao_quadrado(num):
    """Retorna o número ao quadrado."""
    return num ** 2

def ao_cubo(num):
    """Retorna o número ao cubo."""
    return num ** 3

def raiz_quadrada(num):
    """Retorna a raiz quadrada do número."""
    return num ** (1/2)

def raiz_cubica(num):
    """Retorna a raiz cúbica do número."""
    return num ** (1/3)

def main():
    """Função principal que gerencia o menu e a repetição do programa."""
    while True:
        opcao = apresentar_menu()

        if opcao not in [1, 2, 3, 4]:
            print("Saindo da calculadora...")
            break

        numero = float(input("Digite um número: "))

        if opcao == 1:
            resultado = ao_quadrado(numero)
            print(f"O número {numero} ao quadrado é {resultado:.2f}")
        elif opcao == 2:
            resultado = ao_cubo(numero)
            print(f"O número {numero} ao cubo é {resultado:.2f}")
        elif opcao == 3:
            resultado = raiz_quadrada(numero)
            print(f"A raiz quadrada de {numero} é {resultado:.2f}")
        elif opcao == 4:
            resultado = raiz_cubica(numero)
            print(f"A raiz cúbica de {numero} é {resultado:.2f}")

if __name__ == "__main__":
    main()
```

### 7. (Função com retorno com parâmetro) Faça um programa contendo algumas funções:
###a) Crie uma função para apresentar o menu e retornar o **caracter** do cálculo escolhido.

    *** Minha Calculadora ***

        + Soma dois números
        - Subtrai dois números
        * Multiplica dois números
        / Divide dois números
    Qual opção deseja?

###b) Desenvolva uma função para cada opção de cálculo, **mas estas não terão retorno**.

###c) A função main() chamará todas as outras, passando a elas os valores digitados dos dois números para realizar o cálculo.

###Observação: Na chamada da função deve-se utilizar uma estrutura de repetição que permita diversos cálculos, quando o usuário quiser sair da calculadora digitará qualquer valor diferente dos **caracteres** do menu.


```python
def apresentar_menu():
    """Apresenta o menu e retorna o caractere da operação escolhida."""
    print("\n*** Minha Calculadora ***")
    print("    + Soma dois números")
    print("    - Subtrai dois números")
    print("    * Multiplica dois números")
    print("    / Divide dois números")
    print("Para sair digite qualquer outro numero")
    return input("Qual opção deseja? ").strip()

def soma(n1, n2):
    """Realiza a soma de dois números e exibe o resultado."""
    print(f"{n1} + {n2} = {n1 + n2}")

def subtracao(n1, n2):
    """Realiza a subtração de dois números e exibe o resultado."""
    print(f"{n1} - {n2} = {n1 - n2}")

def multiplicacao(n1, n2):
    """Realiza a multiplicação de dois números e exibe o resultado."""
    print(f"{n1} * {n2} = {n1 * n2}")

def divisao(n1, n2):
    """Realiza a divisão de dois números e exibe o resultado."""
    if n2 == 0:
        print("Erro: divisão por zero não permitida!")
    else:
        print(f"{n1} / {n2} = {n1 / n2:.2f}")

def main():
    """Função principal que gerencia o menu e a repetição do programa."""
    while True:
        opcao = apresentar_menu()

        if opcao not in ['+', '-', '*', '/']:
            print("Saindo da calculadora...")
            break

        try:
            n1 = float(input("Digite o primeiro número: "))
            n2 = float(input("Digite o segundo número: "))
        except ValueError:
            print("Erro: Digite valores numéricos válidos.")
            continue

        if opcao == '+':
            soma(n1, n2)
        elif opcao == '-':
            subtracao(n1, n2)
        elif opcao == '*':
            multiplicacao(n1, n2)
        elif opcao == '/':
            divisao(n1, n2)

if __name__ == "__main__":
    main()
```

###9. (Função com retorno com parâmetro) Crie uma função que a partir dos parâmetros/coordenadas (x1, y1) e (x2, y2), calcule e retorne a distância entre estes dois pontos, apresentar na main.distancia.png

```python
# Digite aqui seu código
import math

def calcular_distancia(x1, y1, x2, y2):
    """Calcula e retorna a distância entre dois pontos (x1, y1) e (x2, y2)."""
    distancia = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    return distancia

def main():
    """Função principal para entrada de dados e exibição do resultado."""
    x1 = float(input("Digite a coordenada x1: "))
    y1 = float(input("Digite a coordenada y1: "))
    x2 = float(input("Digite a coordenada x2: "))
    y2 = float(input("Digite a coordenada y2: "))

    distancia = calcular_distancia(x1, y1, x2, y2)

    print(f"A distância entre os pontos ({x1}, {y1}) e ({x2}, {y2}) é {distancia:.2f}")

if __name__ == "__main__":
    main()
```

### 10. (Função com retorno com parâmetro) Crie uma função que verifique se um número é divisível por outro, retorne True ou False e apresente na main. Use o caracter %. Os números são os parâmetros.

```python
# Digite aqui seu código
def eh_divisivel(num1, num2):
    """Verifica se num1 é divisível por num2 e retorna True ou False."""
    if num1 % num2 == 0:
        return True
    else:
        return False

def main():
    """Função principal para entrada e exibição do resultado."""
    num1 = int(input("Digite o primeiro número: "))
    num2 = int(input("Digite o segundo número: "))

    if eh_divisivel(num1, num2) == True:
        print(f"{num1} é divisível por {num2}.")
    else:
        print(f"{num1} não é divisível por {num2}.")

if __name__ == "__main__":
    main()
```

### 11. (Função com retorno com parâmetro) Na função main() o usuário deverá informar a quantidade de linhas e colunas, crie/chame uma função para criar e digitar os elementos de uma matriz, retorne-a. Crie/chame outra função que a calculará a soma de todas as linhas e armazenará num vetor do tamanho da quantidade de linhas, retorne-o e apresente na main().

```python
# Digite aqui seu código. Exemplo de matriz 2x3
def criar_matriz(linhas, colunas):
    """Cria uma matriz com entrada de valores do usuário e a retorna."""
    matriz = []
    for i in range(linhas):
        linha = []
        for j in range(colunas):
            valor = int(input(f"Digite o valor para [{i+1},{j+1}]: "))
            linha.append(valor)
        matriz.append(linha)
    return matriz

def calcular_soma_linhas(matriz):
    """Calcula a soma de cada linha e retorna um vetor com os resultados."""
    return [sum(linha) for linha in matriz]

def main():
    """Função principal."""
    linhas = int(input("Digite o número de linhas: "))
    colunas = int(input("Digite o número de colunas: "))

    matriz = criar_matriz(linhas, colunas)
    soma_linhas = calcular_soma_linhas(matriz)

    print("\nMatriz criada:")
    for linha in matriz:
        print(linha)

    print("\nSoma das linhas:", soma_linhas)

if __name__ == "__main__":
    main()
```

###12. (Função com retorno e com parâmetro). Criar uma função para cadastrar 5 idades, a partir de num vetor que é parâmetro, retorne ele para a main. Crie outra função para calcular a média de idades que terá como parâmetro o vetor de idades, retorne a média e apresente. Chame as duas funções na main.

```python
# Digite aqui seu código
def cadastrar_idades():
    """Cadastra 5 idades em um vetor e o retorna."""
    idades = []
    for i in range(5):
        idade = int(input(f"Digite a idade {i+1}: "))
        idades.append(idade)
    return idades

def calcular_media_idades(idades):
    """Calcula e retorna a média das idades."""
    return sum(idades) / len(idades)

def main():
    """Função principal."""
    idades = cadastrar_idades()
    media = calcular_media_idades(idades)

    print("\nIdades cadastradas:", idades)
    print(f"Média das idades: {media:.2f}")

if __name__ == "__main__":
    main()
```

###14. Crie duas funções:

###a). converter_celsius_para_fahrenheit que recebe um parâmetro: celsius, e retorna a temperatura convertida para Fahrenheit.

###b) converter_temperaturas que recebe uma lista/vetor de temperaturas em Celsius e retorna uma lista/vetor de temperaturas convertidas para Fahrenheit. Esta função deve chamar a função converter_celsius_para_fahrenheit para cada temperatura na lista/vetor.
###Na main crie o vetor, digite 4 temperaturas.
###Chame a função converter_temperaturas.

```python
# Digite aqui seu código
def converter_celsius_para_fahrenheit(celsius):
    """Converte a temperatura de Celsius para Fahrenheit."""
    return (celsius * 9/5) + 32

def converter_temperaturas(lista_celsius):
    """Converte uma lista de temperaturas de Celsius para Fahrenheit."""
    return [converter_celsius_para_fahrenheit(temp) for temp in lista_celsius]

def main():
    """Função principal."""
    temperaturas_celsius = []

    for i in range(4):
        temp = float(input(f"Digite a temperatura {i+1} em Celsius: "))
        temperaturas_celsius.append(temp)

    temperaturas_fahrenheit = converter_temperaturas(temperaturas_celsius)

    print("\nTemperaturas em Celsius:", temperaturas_celsius)
    print("Temperaturas em Fahrenheit:", temperaturas_fahrenheit)

if __name__ == "__main__":
    main()
```
