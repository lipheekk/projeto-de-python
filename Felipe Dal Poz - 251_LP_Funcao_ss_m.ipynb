### 5. (Função sem retorno sem parâmetro) Faça uma função/método que leia um único valor representado em segundos, converta-o e apresente o resultado em horas, minutos e segundos.
    h = segundos / 3600
    r = resto(segundos / 3600)
    m = r / 60
    s = resto(r / 60)
    Observação 1: resto de uma divisão em Python %.
    Observação 2: a hora, o minuto e o segundo devem ser apresentados como números inteiros.

```python
# Digite seu código aqui.
def converter_segundos():
    segundos = int(input("Digite o valor em segundos: "))

    horas = segundos // 3600  # 1 hora = 3600 segundos
    minutos = (segundos % 3600) // 60  # 1 minuto = 60 segundos
    segundos_restantes = segundos % 60  # Restante em segundos

    print(f"{horas} hora(s), {minutos} minuto(s) e {segundos_restantes} segundo(s)")

converter_segundos()
```

### 6. (Função sem retorno sem parâmetro) Faça uma função/método que receba o preço antigo e atual de um produto, determine o percentual de acréscimo entre esses valores e apresente-o.
    r = (100 * preco_novo - 100 * preco_antigo) / preco_antigo

```python
# Digite seu código aqui.
def calcular_acrescimo():
    preco_antigo = float(input("Digite o preço antigo do produto: "))
    preco_atual = float(input("Digite o preço atual do produto: "))

    if preco_antigo == 0:
        print("Não é possível calcular o percentual de acréscimo com preço antigo igual a zero.")
        return

    percentual = ((preco_atual - preco_antigo) / preco_antigo) * 100
    print(f"O percentual de acréscimo foi de {percentual:.2f}%")

calcular_acrescimo()
```

### 8. (Função sem retorno sem parâmetro) Faça uma função/método que leia três notas de um aluno e uma letra, se a letra for igual a *A*, deverá calcular a média aritimética das notas dos alunos, se for *P*, deverá calcular a média ponderada, com pesos 5, 3 e 2. A média deve ser mostrada ao final.
N1, N2 e N3 são notas.

P1, P2 e P3 são pesos.

Média ponderada = (N1 * P1 + N2 * P2 + N3 * P3 ) / (P1 + P2 + P3)

```python
# Digite seu código aqui.
def calcular_media():
    # Lendo as três notas do aluno
    n1 = float(input("Digite a primeira nota: "))
    n2 = float(input("Digite a segunda nota: "))
    n3 = float(input("Digite a terceira nota: "))

    # Lendo o tipo de média
    letra = input("Digite 'A' para média aritmética ou 'P' para média ponderada: ").strip().upper()

    if letra == 'A':
        media = (n1 + n2 + n3) / 3
        print(f"A média aritmética é: {media:.2f}")
    elif letra == 'P':
        # Pesos definidos
        p1, p2, p3 = 5, 3, 2
        media = (n1 * p1 + n2 * p2 + n3 * p3) / (p1 + p2 + p3)
        print(f"A média ponderada é: {media:.2f}")
    else:
        print("Opção inválida. Escolha 'A' ou 'P'.")

# Chamando a função
calcular_media()
```

### 9. (Função sem retorno sem parâmetro) Faça uma função/método que leia uma hora de início e de término de um jogo, ambas divididas em dois valores distintos: hora e minuto. Deverá ser apresentado a duração expressa em minutos, considerando que o tempo máximo de duração de um jogo é de 24 horas e que ele pode começar em um dia e terminar no outro.
    se m_f < m_i
        m_f = m_f + 60
        h_f = h_f - 1
    se h_f < h_i
        h_f = h_f + 24
    tot_m = m_f - m_i
    tot_h = h_f - h_i
    total = tot_h * 60 + tot_m

```python
# Digite seu código aqui.
def calcular_duracao_jogo():
    # Solicita a entrada do usuário
    h_i = int(input("Informe a hora de início do jogo (0-23): "))
    m_i = int(input("Informe os minutos de início do jogo (0-59): "))
    h_f = int(input("Informe a hora de término do jogo (0-23): "))
    m_f = int(input("Informe os minutos de término do jogo (0-59): "))
    if m_f < m_i:
        m_f += 60
        h_f -= 1

    if h_f < h_i:
        h_f += 24

    tot_m = m_f - m_i
    tot_h = h_f - h_i
    total = tot_h * 60 + tot_m
    print(f"A duração total do jogo foi de {total} minutos.")

calcular_duracao_jogo()
```

### 11. (Função sem retorno sem parâmetro) Faça uma função/método que leia um valor inteiro e positivo N e mostre o valor de S, obtido pelo seguinte cálculo:
###S = 1  +  1/1!  +  1/2!  +  1/3!  + ... +  1/N!

Observvação: Não pode usar vetor e função pronta da linguagem Python.

```python
# Resultado dos fatoriais:
# 1! = 1                     1/1! => 1/1  = 1
# 2! = 2                     1/2! => 1/2  = 0.5
# 3! = 3 * 2 * 1 = 6         1/3! => 1/6  = 0.16
# 4! = 1 * 2 * 3 * 4 = 24    1/4! => 1/24 = 0.041
# S = 1
# S = 1 + 1 + 1/2! + 1/3 + 1/4!
# S = 1 + 1 + 0.5 + 0.16 + 0.041
# S = 2.708

# N = int(input('Informe o valor de N termos: '))
def calcular_s():
    N = int(input("Informe um valor inteiro e positivo para N: "))

    if N < 0:
        print("O valor deve ser um número inteiro positivo.")
        return

    S = 1
    fatorial = 1

    for i in range(1, N + 1):
        fatorial *= i
        S += 1 / fatorial
    print(f"O valor de S é: {S:.6f}")

calcular_s()

```
