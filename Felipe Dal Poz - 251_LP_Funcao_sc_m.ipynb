### 9. (Função sem retorno com parâmetro) Faça um programa que a partir da função main(), o usuário escolha quais operações aritméticas (***) quer realizar, para isso coloque as chamadas das funções dentro de uma estrutura de repetição. Não use NENHUMA função pronta da linguagem Python.
Exemplo:

..............................................

Programa para cálculos

..............................................

1 - Soma

2 - Subtração

3 - Multiplicação

4 - Divisão

5 - Potência

6 - Raiz

7 - Sair


```python
def soma(a, b):
    """Realiza a soma de dois números."""
    resultado = a + b
    print(f"Resultado da soma: {resultado}")

def subtracao(a, b):
    """Realiza a subtração de dois números."""
    resultado = a - b
    print(f"Resultado da subtração: {resultado}")

def multiplicacao(a, b):
    """Realiza a multiplicação de dois números."""
    resultado = a * b
    print(f"Resultado da multiplicação: {resultado}")

def divisao(a, b):
    """Realiza a divisão de dois números, sem permitir divisão por zero."""
    if b == 0:
        print("Erro: Divisão por zero não é permitida.")
    else:
        resultado = a / b
        print(f"Resultado da divisão: {resultado}")

def potencia(base, expoente):
    """Calcula a potência de um número sem usar a função pow()."""
    resultado = 1
    for _ in range(expoente):
        resultado *= base
    print(f"Resultado da potência: {resultado}")

def raiz(numero):
    """Calcula a raiz quadrada de um número usando um método iterativo."""
    if numero < 0:
        print("Erro: Não é possível calcular a raiz de um número negativo.")
        return

    x = numero
    y = (x + 1) / 2  # Estimativa inicial
    while abs(x - y) > 0.0001:
        x = y
        y = (x + numero / x) / 2  # Método de Newton

    print(f"Resultado da raiz quadrada: {x}")

def main():
    while True:
        print("\n" + "." * 50)
        print("Programa para cálculos".center(50))
        print("." * 50)
        print("1 - Soma")
        print("2 - Subtração")
        print("3 - Multiplicação")
        print("4 - Divisão")
        print("5 - Potência")
        print("6 - Raiz")
        print("7 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "7":
            print("Encerrado o programa...")
            break

        if opcao in ["1", "2", "3", "4", "5"]:
            num1 = int(input("Digite o primeiro número: "))
            num2 = int(input("Digite o segundo número: "))

            if opcao == "1":
                soma(num1, num2)
            elif opcao == "2":
                subtracao(num1, num2)
            elif opcao == "3":
                multiplicacao(num1, num2)
            elif opcao == "4":
                divisao(num1, num2)
            elif opcao == "5":
                potencia(num1, num2)

        elif opcao == "6":
            num = int(input("Digite o número para calcular a raiz quadrada: "))
            raiz(num)

        else:
            print("Opção inválida! Escolha um número de 1 a 7.")

main()
```

```python
# Definindo os vetores de dados
vetProduto = []
vetQtde = []
vetPreco = []

# Função 1: Cadastro do nome, quantidade e preço do produto
def cadastrar_produto():
    for i in range(10):
        nome = input(f"Digite o nome do produto {i + 1}: ")
        quantidade = int(input(f"Digite a quantidade do produto {i + 1}: "))
        preco = float(input(f"Digite o preço do produto {i + 1}: R$ "))
        vetProduto.append(nome)
        vetQtde.append(quantidade)
        vetPreco.append(preco)

# Função 2: Apresentação de todos os produtos
def apresentar_produtos():
    print("{:<30} {:>10} {:>15}".format("Nome", "Quantidade", "Preço R$"))
    print("-" * 55)
    for i in range(len(vetProduto)):
        print("{:<30} {:>10} {:>15.2f}".format(vetProduto[i], vetQtde[i], vetPreco[i]))

# Função 3: Apresentação de estoque mínimo (quantidade <= 3)
def apresentar_estoque_minimo():
    print("{:<30} {:>10} {:>15}".format("Nome", "Quantidade", "Preço R$"))
    print("-" * 55)
    for i in range(len(vetProduto)):
        if vetQtde[i] <= 3:
            print("{:<30} {:>10} {:>15.2f}".format(vetProduto[i], vetQtde[i], vetPreco[i]))

# Função 4: Aplicação de reajuste de aumento de 5% para produtos começando com "C"
def reajustar_aumento():
    for i in range(len(vetProduto)):
        if vetProduto[i].lower().startswith('c'):
            vetPreco[i] *= 1.05

# Função 5: Alteração de nome de produto de "camisa" para "blusa"
def alterar_nome_camisa():
    for i in range(len(vetProduto)):
        vetProduto[i] = vetProduto[i].replace("Camisa", "Blusa")

# Função 6: Capitalize todos os nomes de produtos
def capitalize_nomes():
    for i in range(len(vetProduto)):
        vetProduto[i] = vetProduto[i].capitalize()

# Função para exibir o menu e capturar a opção do usuário
def exibir_menu():
    while True:
        print("\nPrograma para Gerenciamento de Estoque de Produtos")
        print("1 - Cadastro do nome, quantidade e preço do produto")
        print("2 - Apresentação de todos os produtos")
        print("3 - Apresentação de estoque mínimo")
        print("4 - Aplicação de reajuste de aumento 5%")
        print("5 - Alteração de nome de produto de camisa para blusa")
        print("6 - Capitalize todos os nomes de produtos")
        print("7 - Sair")

        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            cadastrar_produto()
        elif opcao == 2:
            apresentar_produtos()
        elif opcao == 3:
            apresentar_estoque_minimo()
        elif opcao == 4:
            reajustar_aumento()
        elif opcao == 5:
            alterar_nome_camisa()
        elif opcao == 6:
            capitalize_nomes()
        elif opcao == 7:
            print("Saindo do programa...")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Chamar a função para exibir o menu
exibir_menu()

```

```python
def calcular_faturamento_total(shoppings, faturamentos):
    faturamento_total = {}

    for i in range(len(shoppings)):
        if shoppings[i] in faturamento_total:
            faturamento_total[shoppings[i]] += faturamentos[i]
        else:
            faturamento_total[shoppings[i]] = faturamentos[i]

    print("\nFaturamento total de cada shopping:")
    for shopping, total in faturamento_total.items():
        print(f"{shopping}: R$ {total:.2f}")

# Função principal
if __name__ == "__main__":
    shoppings = []
    faturamentos = []

    for i in range(10):
        nome = input(f"Digite o nome do shopping {i+1}: ")
        faturamento = float(input(f"Digite o faturamento do shopping {nome}: "))
        shoppings.append(nome)
        faturamentos.append(faturamento)

    calcular_faturamento_total(shoppings, faturamentos)
```

###22. De acordo com a figura a seguir crie uma função que valide a senha, seguindo as regras demonstradas. Digite a senha na função main(), chame função de validação e apresente se a senha está correta ou incorreta.
senha.png

```python
def valida_senha(senha):
    """
    Valida a senha de acordo com as regras (versão "bruta").
    """
    tem_numero = False
    tem_minusculo = False
    tem_maiusculo = False
    tem_especial = False

    if len(senha) < 8:
        return False  # Senha muito curta

    for caractere in senha:
        if caractere.isdigit():
            tem_numero = True
        elif caractere.islower():
            tem_minusculo = True
        elif caractere.isupper():
            tem_maiusculo = True
        elif caractere in "!@#&()-[[]]]:;','\"?/*~$^+=<>._]%)":
            tem_especial = True

    if tem_numero and tem_minusculo and tem_maiusculo and tem_especial:
        return True  # Senha válida
    else:
        return False  # Senha inválida

if __name__ == "__main__":
    senha = input("Digite a senha: ")
    if valida_senha(senha):
        print("Senha válida!")
    else:
        print("Senha inválida! Por favor, siga as regras.")
```
