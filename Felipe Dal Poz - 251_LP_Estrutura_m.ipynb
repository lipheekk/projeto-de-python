```python
# CRUD - Create, Read, Update, Delete - Inserção, Visualização, Atualização, Exclusão
class TipoProduto: # tipo de dado heterogêneo (int, str, float)
    codigo = 0 # codigo é int
    nome = '' # nome é str
    preco = 0.0 # preco é float

def menu():
    print('\nMenu do Sistema')
    print('1 - Cadastrar')
    print('2 - Reajustar')
    print('3 - Visualizar')
    print('4 - Consultar por nome - cont')
    print('5 - Consultar por nome - bool')
    print('6 - Consultar por nome - if in')
    print('7 - Alterar nome e preço a partir do código')
    print('8 - Excluir a produto com menor preço')
    print('9 - Sair')
    return int(input('Qual opção deseja? '))

def cadastrar(vprod):# entra vazio
    for i in range(2):
        p = TipoProduto() #inicialização de variável
        p.codigo = int(input('\nDigite o código: '))
        p.nome = input('Digite o nome: ')
        p.preco = float(input('Digite o preço: '))
        vprod.append(p)
    return vprod # sai com produtos cadastrados

def visualizar(vprod): # vprod é um PARÂMETRO
    if len(vprod) > 0:
        for i in range(len(vprod)): # < 2   i=0
            print('Código: ',vprod[i].codigo,'\tNome:',vprod[i].nome,'\tPreço R$',vprod[i].preco)
    else:
        print('\nNão há produtos cadastrados!')

def reajustar(vprod): # 10 %
    if len(vprod) > 0:
        for i in range(len(vprod)):
            vprod[i].preco += vprod[i].preco * 10 / 100
        print('\nTodos os produtos foram reajustados!')
    else:
        print('\nNão há produtos cadastrados!')
    return vprod

def consulta_nome_cont(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do produto que quer pesquisar: ')
        cont = 0 # inicialize as variáveis contadoras ou somatórias antes de
        # entrar na estrutura de repetição
        for i in range(len(vprod)):
            if nome_pesquisar == vprod[i].nome:
                cont += 1 # cont = cont + 1
        if cont >= 1:
            print('\nEncontrei o produto!')
        else:
            print('\nNão encontrei o produto!')
    else:
        print('\nNão há produtos cadastrados!')

def consulta_nome_bool(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do produto que quer pesquisar: ')
        achei_produto = False
        for i in range(len(vprod)):
            if nome_pesquisar == vprod[i].nome:
                achei_produto = True
        if achei_produto: # if achei_produto == True:
            print('\nEncontrei o produto!')
        else:
            print('\nNão encontrei o produto!')
    else:
        print('\nNão há produtos cadastrados!')

def consulta_nome_if_in(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do produto que quer pesquisar: ')
        achei_produto = False
        for produto in vprod: # foreach
            if nome_pesquisar in produto.nome:
                achei_produto = True # ou usar o cont = cont + 1
        if achei_produto: # if achei_produto == True:
            print('\nEncontrei o produto!')
        else:
            print('\nNão encontrei o produto!')
    else:
        print('\nNão há produtos cadastrados!')

def alterar(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        codigo_pesquisar = int(input('\nInforme o código do produto que quer pesquisar: '))
        achei_produto = False
        for produto in vprod: # foreach
            if codigo_pesquisar == produto.codigo:
                # momento de realizar a alteração
                produto.nome = input('Digite o nome a alterar: ')
                produto.preco = float(input('Digite o preço a alterar: '))
                achei_produto = True # ou usar o cont = cont + 1
                print('\nAlteração realizada com sucesso!')
        if not achei_produto:
            print('\nNão há produto cadastrado com esse código!')
    else:
        print('\nNão há produtos cadastrados!')
    return vprod

def excluir(vprod):
    if len(vprod) > 0:
        for i in range(len(vprod)):
            if i == 0: # esse código apenas executa uma vez
                menor_preco = vprod[i].preco
                indice = i
            if vprod[i].preco < menor_preco:
                menor_preco = vprod[i].preco
                indice = i
        vprod.pop(indice)
        print('\nExclusão realizada com sucesso!')
    else:
        print('\nNão há produtos cadastrados!')
    return vprod

def main():
    vet_produto = [] # vetor vazio
    opcao = menu()
    while opcao >= 1 and opcao <= 8:
        if opcao == 1: # vet_produto entra vazio
            #voltar com conteúdo cadastrado
            vet_produto = cadastrar(vet_produto)
        elif opcao == 2:
            vet_produto = reajustar(vet_produto)
        elif opcao == 3:
            visualizar(vet_produto)
        elif opcao == 4: # quando chama a função a variável que esta dentro do
            consulta_nome_cont(vet_produto) # parenteses é ARGUMENTO = vet_produto
        elif opcao == 5:
            consulta_nome_bool(vet_produto)
        elif opcao == 6:
            consulta_nome_if_in(vet_produto)
        elif opcao == 7:
            vet_produto = alterar(vet_produto)
        elif opcao == 8:
            vet_produto = excluir(vet_produto)
        opcao = menu() # última linha de código do while, faça o input novamente

main()
```

```python
# Percorra todos os vetor e quando a condição de
# enontrar um menor saldo for executada, guarde o índice
# Fora do laço de repetição, exclua pelo índice,
# por exemplo com o comando pop(índice)
# Digite seu código aqui
# CRUD - Create, Read, Update, Delete - Inserção, Visualização, Atualização, Exclusão
class TipoProduto: # tipo de dado heterogêneo (int, str, float)# codigo é int # nome é str # preco é float
    nome = ''
    nconta = 0
    saldo = 0.0


def menu():
    print('\nMenu do Sistema')
    print('1 - Cadastrar contas')
    print('2 - Visualizar')
    print('3 - Consultar por nome - cont')
    print('4 - Consultar por nome - bool')
    print('5 - Consultar por nome - if in')
    print('6 - Alterar contas')
    print('7 - Excluir a conta com menor saldo')
    print('8 - Sair')
    return int(input('Qual opção deseja? '))

def cadastrar(vprod):# entra vazio
    for i in range(2):
        p = TipoProduto() #inicialização de variável
        #p. = int(input('\nDigite o código: '))
        p.nome = input('\nDigite o nome: ')
        p.nconta = int(input('Digite o codigo de numeros da sua conta: '))
        p.saldo = float(input('Digite seu saldo: '))
        vprod.append(p)
    return vprod # sai com produtos cadastrados

def visualizar(vprod): # vprod é um PARÂMETRO
    if len(vprod) > 0:
        for i in range(len(vprod)): # < 2   i=0
            print('Nome do Titular: ',vprod[i].nome,'\tNumero da conta:',vprod[i].nconta,'\tSaldo:',vprod[i].saldo)
    else:
        print('\nNão há contas cadastrados!')


def consulta_nome_cont(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do titular que quer pesquisar: ')
        cont = 0 # inicialize as variáveis contadoras ou somatórias antes de
        # entrar na estrutura de repetição
        for i in range(len(vprod)):
            if nome_pesquisar == vprod[i].nome:
                cont += 1 # cont = cont + 1
        if cont >= 1:
            print('\nEncontrei a conta!')
        else:
            print('\nNão encontrei a conta!')
    else:
        print('\nNão há contas para encontrar!')

def consulta_nome_bool(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do titular que quer pesquisar: ')
        achei_produto = False
        for i in range(len(vprod)):
            if nome_pesquisar == vprod[i].nome:
                achei_produto = True
        if achei_produto: # if achei_produto == True:
            print('\nEncontrei a conta!')
        else:
            print('\nNão encontrei a conta!')
    else:
        print('\nNão há contas cadastrados!')

def consulta_nome_if_in(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        nome_pesquisar = input('\nInforme o nome do titular que quer pesquisar: ')
        achei_produto = False
        for produto in vprod: # foreach
            if nome_pesquisar in produto.nome:
                achei_produto = True # ou usar o cont = cont + 1
        if achei_produto: # if achei_produto == True:
            print('\nEncontrei a conta!')
        else:
            print('\nNão encontrei a conta!')
    else:
        print('\nNão há contas cadastrados!')
def alterar(vprod):
    if len(vprod) > 0: # tem produtos (elementos) no vetor
        codigo_pesquisar = int(input('\nInforme o código da conta que quer alterar: '))
        achei_produto = False
        for produto in vprod: # foreach
            if codigo_pesquisar == produto.nconta:
                # momento de realizar a alteração
                produto.nome = input('Digite o nome a alterar: ')
                produto.saldo = float(input('Digite o preço a alterar: '))
                achei_produto = True # ou usar o cont = cont + 1
                print('\nAlteração realizada com sucesso!')
        if not achei_produto:
            print('\nNão há contas cadastradas com esse código!')
    else:
        print('\nNão há contas cadastradas!')
    return vprod

def excluir(vprod):
    if len(vprod) > 0:
        for i in range(len(vprod)):
            if i == 0: # esse código apenas executa uma vez
                menor_saldo = vprod[i].saldo
                indice = i
            if vprod[i].saldo < menor_saldo:
                menor_saldo = vprod[i].saldo
                indice = i
        vprod.pop(indice)
        print('\nExclusão realizada com sucesso!')
    else:
        print('\nNão há contas cadastrados!')
    return vprod


def main():
    vet_produto = [] # vetor vazio
    opcao = menu()
    while opcao >= 1 and opcao <= 7:
        if opcao == 1: # vet_produto entra vazio
            #voltar com conteúdo cadastrado
            vet_produto = cadastrar(vet_produto)
        elif opcao == 2:
            visualizar(vet_produto)
        elif opcao == 3: # quando chama a função a variável que esta dentro do
            consulta_nome_cont(vet_produto) # parenteses é ARGUMENTO = vet_produto
        elif opcao == 4:
            consulta_nome_bool(vet_produto)
        elif opcao == 5:
            consulta_nome_if_in(vet_produto)
        elif opcao == 6:
            vet_produto = alterar(vet_produto)
        elif opcao == 7:
            vet_produto = excluir(vet_produto)
        opcao = menu()

main()
```

```python
# Por onde começo a programar esse exercício?
# Sempre comece da parte mais simples/talvez a menor e já
# vai implementando e testando cada uma das funções
# Programe o cadastro e o relatório para testar e assim até a
# última função.
class TipoCliente:
    cod_cli = 0
    nome = ''
    fone = ''


class TipoDocumento:
    num_doc = 0
    cod_cli = 0
    dia_venc = 0
    dia_pag = 0
    valor = 0.0
    juros = 0.0


clientes = []
documentos = []


def menu():
    print('\n\n*** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS ***')
    print('1 - Cadastrar clientes')
    print('2 - Relatório de clientes')
    print('3 - Cadastrar documentos')
    print('4 - Relatório de documentos')
    print('5 - Excluir clientes sem documentos')
    print('6 - Excluir documentos pelo número')
    print('7 - Excluir documentos por cliente')
    print('8 - Excluir documentos por período')
    print('9 - Alterar informações dos clientes')
    print('10 - Mostrar total de documentos de um cliente')
    print('11 - Sair')
    return int(input('Qual opção deseja? '))


def cadastrar_cliente():
    if len(clientes) >= 15:
        print('Limite máximo de clientes atingido.')
        return

    cod_cli = int(input('Código do cliente: '))
    for cli in clientes:
        if cli.cod_cli == cod_cli:
            print('Cliente já cadastrado com esse código!')
            return

    nome = input('Nome do cliente: ')
    fone = input('Telefone do cliente: ')

    novo_cliente = TipoCliente()
    novo_cliente.cod_cli = cod_cli
    novo_cliente.nome = nome
    novo_cliente.fone = fone

    clientes.append(novo_cliente)
    print('Cliente cadastrado com sucesso.')


def relatorio_clientes():
    if not clientes:
        print('Nenhum cliente cadastrado.')
        return

    print('\n--- RELATÓRIO DE CLIENTES ---')
    for cli in clientes:
        print(f'Código: {cli.cod_cli} | Nome: {cli.nome} | Fone: {cli.fone}')

def cadastrar_documento():
    if len(documentos) >= 30:
        print('Limite máximo de documentos atingido.')
        return

    num_doc = int(input('Número do documento: '))
    for doc in documentos:
        if doc.num_doc == num_doc:
            print('Documento já cadastrado com esse número!')
            return

    cod_cli = int(input('Código do cliente: '))
    cliente_encontrado = False
    for cli in clientes:
        if cli.cod_cli == cod_cli:
            cliente_encontrado = True
            break

    if not cliente_encontrado:
        print('Cliente não encontrado. Documento não cadastrado.')
        return

    dia_venc = int(input('Dia de vencimento: '))
    dia_pag = int(input('Dia de pagamento: '))
    valor = float(input('Valor do documento: '))

    novo_doc = TipoDocumento()
    novo_doc.num_doc = num_doc
    novo_doc.cod_cli = cod_cli
    novo_doc.dia_venc = dia_venc
    novo_doc.dia_pag = dia_pag
    novo_doc.valor = valor

    if dia_pag > dia_venc:
        novo_doc.juros = valor * 0.05
    else:
        novo_doc.juros = 0.0

    documentos.append(novo_doc)
    print('Documento cadastrado com sucesso.')


def relatorio_documentos():
    if not documentos:
        print('Nenhum documento cadastrado.')
        return

    print('\n--- RELATÓRIO DE DOCUMENTOS ---')
    for doc in documentos:
        print(f'Número: {doc.num_doc} | Cliente: {doc.cod_cli} | '
              f'Vencimento: {doc.dia_venc} | Pagamento: {doc.dia_pag} | '
              f'Valor: {doc.valor:.2f} | Juros: {doc.juros:.2f}')


def excluir_clientes_sem_documentos():
    excluidos = 0
    for cli in clientes[:]:
        tem_documento = False
        for doc in documentos:
            if doc.cod_cli == cli.cod_cli:
                tem_documento = True
                break
        if not tem_documento:
            clientes.remove(cli)
            excluidos += 1
    print(f'{excluidos} cliente(s) excluído(s) sem documentos.')


def excluir_documento_por_numero():
    num_doc = int(input('Número do documento a excluir: '))
    for doc in documentos:
        if doc.num_doc == num_doc:
            documentos.remove(doc)
            print('Documento excluído com sucesso.')
            return
    print('Documento não encontrado.')


def excluir_documentos_por_cliente():
    cod_cli = int(input('Código do cliente: '))
    cliente_existe = any(cli.cod_cli == cod_cli for cli in clientes)

    if not cliente_existe:
        print('Cliente não encontrado.')
        return

    removidos = 0
    for doc in documentos[:]:
        if doc.cod_cli == cod_cli:
            documentos.remove(doc)
            removidos += 1

    print(f'{removidos} documento(s) excluído(s) do cliente {cod_cli}.')


def excluir_documentos_por_periodo():
    dia_inicial = int(input('Dia inicial: '))
    dia_final = int(input('Dia final: '))
    removidos = 0

    for doc in documentos[:]:
        if dia_inicial <= doc.dia_venc <= dia_final:
            documentos.remove(doc)
            removidos += 1

    print(f'{removidos} documento(s) excluído(s) no período de vencimento informado.')


# 9 - Alterar informações dos clientes
def alterar_cliente():
    cod_cli = int(input('Código do cliente a alterar: '))
    for cli in clientes:
        if cli.cod_cli == cod_cli:
            print('Deixe vazio se não quiser alterar.')
            novo_nome = input(f'Nome atual ({cli.nome}): ')
            novo_fone = input(f'Fone atual ({cli.fone}): ')

            if novo_nome.strip() != '':
                cli.nome = novo_nome
            if novo_fone.strip() != '':
                cli.fone = novo_fone

            print('Informações atualizadas com sucesso.')
            return
    print('Cliente não encontrado.')


# 10 - Mostrar total de documentos de um cliente
def total_documentos_cliente():
    cod_cli = int(input('Código do cliente: '))
    total = 0
    for doc in documentos:
        if doc.cod_cli == cod_cli:
            total += 1

    print(f'O cliente {cod_cli} possui {total} documento(s).')


# ===== Loop Principal =====

while True:
    opcao = menu()

    if opcao == 1:
        cadastrar_cliente()
    elif opcao == 2:
        relatorio_clientes()
    elif opcao == 3:
        cadastrar_documento()
    elif opcao == 4:
        relatorio_documentos()
    elif opcao == 5:
        excluir_clientes_sem_documentos()
    elif opcao == 6:
        excluir_documento_por_numero()
    elif opcao == 7:
        excluir_documentos_por_cliente()
    elif opcao == 8:
        excluir_documentos_por_periodo()
    elif opcao == 9:
        alterar_cliente()
    elif opcao == 10:
        total_documentos_cliente()
    elif opcao == 11:
        print('Encerrando o sistema...')
        break
    else:
        print('Opção inválida. Tente novamente.')

```

Lembre-se de que a empresa só pode prestar os serviços que já tenham sido cadastrados no vetor de tipo de serviços.

Caso o usuário digite um código de tipo de serviço inválido, o programa deverá mostrar uma mensagem de erro.

Quando o usuário tentar cadastrar mais de três serviços prestados em um mesmo dia, também deverá mostrar uma mensagem de erro.

**Para a opção 5**: o programa deverá receber o dia que se deseja consultar e mostrar os respectivos serviços prestados.

**Para a opção 6**: o programa deverá receber o valor mínimo e o valor máximo e mostrar os serviços prestados que estiverem nesse intervalo.

**Para a opção 7**: o programa deverá mostrar todos os serviços prestados, conforme o exemplo a seguir.

                  DIA 01


No do serviço| Valor do serviço R$| Código do serviço| Descrição| Código do cliente
---|---|---|---|---
100| 200,00| 1| Pintura| 1
150| 100,00| 3| Faxina| 5
201| 640,00| 4| Reforma| 2

                  DIA 02

No do serviço| Valor do serviço R$| Código do serviço| Descrição| Código do cliente
---|---|---|---|---
301| 600,00| 4| Reforma| 3
280| 352,00| 1| Pintura| 2
306| 200,00| 2| Jardinagem| 1

```python
# Digite seu código aqui
tipo_servico = []

servicos_prestados = [[None for _ in range(3)] for _ in range(30)]

def cadastrar_tipos_servicos():
    tipo_servico.clear()
    for i in range(4):
        codigo = int(input(f"Digite o código do tipo de serviço {i+1}: "))
        descricao = input(f"Digite a descrição do serviço {i+1}: ")
        tipo_servico.append({'codigo': codigo, 'descricao': descricao})
    print("Tipos de serviço cadastrados com sucesso!")

def mostrar_tipos_servico():
    if not tipo_servico:
        print("Nenhum tipo de serviço cadastrado.")
        return
    print("Código\tDescrição")
    for serv in tipo_servico:
        print(f"{serv['codigo']}\t{serv['descricao']}")

def obter_descricao_por_codigo(codigo):
    for serv in tipo_servico:
        if serv['codigo'] == codigo:
            return serv['descricao']
    return None

def cadastrar_servico_prestado():
    dia = int(input("Digite o dia (1-30): "))
    if dia < 1 or dia > 30:
        print("Dia inválido.")
        return
    dia_index = dia - 1
    dia_servicos = servicos_prestados[dia_index]

    if all(s is not None for s in dia_servicos):
        print("Limite de 3 serviços para esse dia já foi atingido.")
        return

    num_servico = int(input("Número do serviço: "))
    valor = float(input("Valor do serviço: R$ "))
    cod_servico = int(input("Código do serviço: "))
    cod_cliente = int(input("Código do cliente: "))

    if obter_descricao_por_codigo(cod_servico) is None:
        print("Código de serviço inválido.")
        return

    for i in range(3):
        if dia_servicos[i] is None:
            dia_servicos[i] = {
                'numero': num_servico,
                'valor': valor,
                'codigo_servico': cod_servico,
                'codigo_cliente': cod_cliente
            }
            print("Serviço cadastrado com sucesso.")
            return

def mostrar_todos_servicos():
    print("SERVIÇOS PRESTADOS:")
    for dia in range(30):
        if any(servicos_prestados[dia]):
            print(f"\nDIA {dia + 1}")
            print("No do serviço\tValor R$\tCódigo Serviço\tDescrição\tCódigo Cliente")
            for servico in servicos_prestados[dia]:
                if servico:
                    desc = obter_descricao_por_codigo(servico['codigo_servico'])
                    print(f"{servico['numero']}\t\t{servico['valor']:.2f}\t\t{servico['codigo_servico']}\t\t{desc}\t\t{servico['codigo_cliente']}")

def mostrar_servicos_por_dia():
    dia = int(input("Digite o dia que deseja consultar (1-30): "))
    if dia < 1 or dia > 30:
        print("Dia inválido.")
        return
    dia_index = dia - 1
    servicos = servicos_prestados[dia_index]
    if not any(servicos):
        print("Nenhum serviço prestado nesse dia.")
        return
    print(f"\nDIA {dia}")
    print("No do serviço\tValor R$\tCódigo Serviço\tDescrição\tCódigo Cliente")
    for servico in servicos:
        if servico:
            desc = obter_descricao_por_codigo(servico['codigo_servico'])
            print(f"{servico['numero']}\t\t{servico['valor']:.2f}\t\t{servico['codigo_servico']}\t\t{desc}\t\t{servico['codigo_cliente']}")

def mostrar_servicos_por_valor():
    min_valor = float(input("Digite o valor mínimo: "))
    max_valor = float(input("Digite o valor máximo: "))
    print("\nServiços no intervalo especificado:")
    achou = False
    for dia in range(30):
        for servico in servicos_prestados[dia]:
            if servico and min_valor <= servico['valor'] <= max_valor:
                desc = obter_descricao_por_codigo(servico['codigo_servico'])
                print(f"DIA {dia + 1} - Serviço: {servico['numero']}, Valor: R$ {servico['valor']:.2f}, Código: {servico['codigo_servico']}, Descrição: {desc}, Cliente: {servico['codigo_cliente']}")
                achou = True
    if not achou:
        print("Nenhum serviço encontrado no intervalo de valores.")

def relatorio_geral():
    print("\nRELATÓRIO GERAL")
    for dia in range(30):
        if any(servicos_prestados[dia]):
            print(f"\nDIA {dia + 1}")
            print("No do serviço\tValor R$\tCódigo Serviço\tDescrição\tCódigo Cliente")
            for servico in servicos_prestados[dia]:
                if servico:
                    desc = obter_descricao_por_codigo(servico['codigo_servico'])
                    print(f"{servico['numero']}\t\t{servico['valor']:.2f}\t\t{servico['codigo_servico']}\t\t{desc}\t\t{servico['codigo_cliente']}")

# Menu principal
def menu():
    while True:
        print("\nMENU DE OPÇÕES:")
        print("1 - Cadastrar os tipos de serviços")
        print("2 - Mostrar todos os tipos de serviço")
        print("3 - Cadastrar os serviços prestados")
        print("4 - Mostrar todos os serviços prestados")
        print("5 - Mostrar os serviços prestados em determinado dia")
        print("6 - Mostrar os serviços prestados dentro de um intervalo de valor")
        print("7 - Mostrar um relatório geral (por dia, com descrição do tipo)")
        print("8 - Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            cadastrar_tipos_servicos()
        elif opcao == '2':
            mostrar_tipos_servico()
        elif opcao == '3':
            cadastrar_servico_prestado()
        elif opcao == '4':
            mostrar_todos_servicos()
        elif opcao == '5':
            mostrar_servicos_por_dia()
        elif opcao == '6':
            mostrar_servicos_por_valor()
        elif opcao == '7':
            relatorio_geral()
        elif opcao == '8':
            print("Encerrando o programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Início do programa
menu()
```
