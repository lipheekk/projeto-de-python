###7. (Função com retorno sem parâmetro) Faça um programa contendo uma função/método que retorna a quantidade de letras numa palavra que deve ser lida. Faça a chamada da função na main(). Não use função pronta do Python.

```python
# Digite aqui seu código
def contar_letras(palavra):
    # Variável para contar as letras
    contador = 0

    # Percorre cada caractere da palavra
    for letra in palavra:
        contador += 1  # Incrementa o contador a cada letra

    # Retorna a quantidade de letras
    return contador

def main():
    # Lê a palavra do usuário
    palavra = input("Digite uma palavra: ")

    # Chama a função para contar as letras da palavra
    quantidade = contar_letras(palavra)

    # Exibe o resultado
    print(f"A quantidade de letras na palavra '{palavra}' é: {quantidade}")

# Chama a função principal
main()
```

###8.  (Função com retorno sem parâmetro) Faça um programa contendo uma função/método que retorna o maior número de uma lista. Faça a chamada na função main().

```python
def encontrar_maior():
    lista = []  # Lista para armazenar os números
    maior = float('-inf')  # Inicializa a variável maior com o menor valor possível

    # Solicita ao usuário o número de membros na lista
    n = int(input('Número de membros na lista: '))

    # Solicita ao usuário que digite os números
    for i in range(n):
        numero = int(input(f"Digite o {i + 1}º número: "))  # Lê o número do usuário
        lista.append(numero)  # Adiciona o número à lista

        # Verifica se o número é maior que o maior encontrado até agora
        if numero > maior:
            maior = numero

    return maior  # Retorna o maior número encontrado

def main():
    # Chama a função para encontrar o maior número da lista
    maior_numero = encontrar_maior()

    # Exibe o maior número encontrado
    print(f"O maior número da lista é: {maior_numero}")

# Chama a função principal
main()
```

```python
import datetime

def verificar_fim_de_semana():
    # Obtém o dia atual da semana (0=segunda, 6=domingo)
    dia_atual = datetime.datetime.now().weekday()

    if dia_atual == 5 or dia_atual == 6:
        return True
    else:
        return False

def main():

    if verificar_fim_de_semana():
        print("Hoje é fim de semana!")
    else:
        print("Hoje não é fim de semana.")


main()
```

###15. (Função com retorno sem parâmetro) Faça uma função/método que leia e armazene 5 elementos inteiros no vetor A, deverá ser gerado um vetor B, de mesmo tamanho, que armazenará o fatorial de cada elemento de A. Não use função pronta de cálculo de fatorial. Retorne/apresente o vetor B.

```python
# Exemplo:
# A = [5  ,4 ,3,2,1]
# B = [120,24,6,2,1]
```

```python
def calcular_fatorial(n):

    fatorial = 1
    for i in range(1, n + 1):
        fatorial *= i
    return fatorial

def gerar_vetor_b(a):

    b = []
    for numero in a:
        b.append(calcular_fatorial(numero))
    return b

def main():
    A = []
    print("Digite 5 números inteiros:")
    for i in range(5):
        A.append(int(input(f"Digite o {i + 1}º número: ")))

    B = gerar_vetor_b(A)

    print("Vetor A:", A)
    print("Vetor B (fatoriais):", B)

main()
```

```python
def encontrar_numeros_perfeitos():
    numeros_perfeitos = []
    n = 1

    while len(numeros_perfeitos) < 3:
        soma_divisores = 0

        for i in range(1, n):
            if n % i == 0:
                soma_divisores += i

        if soma_divisores == n:
            numeros_perfeitos.append(n)

        n += 1

    return numeros_perfeitos

def main():
    perfeitos = encontrar_numeros_perfeitos()

    print(f"Os três primeiros números perfeitos são: {perfeitos}")

main()
```

```python
idade = []  # Lista global para armazenar as idades
genero = []  # Lista global para armazenar os gêneros
olho = []  # Lista global para armazenar as cores dos olhos
cabelo = []  # Lista global para armazenar as cores dos cabelos

def cadastrar():
    # Função que coleta e armazena os dados dos habitantes
    for i in range(100):
        print(f"Cadastro do {i+1}º habitante:")
        idade_valor = int(input("Idade: "))
        genero_valor = input("Gênero (M/F): ").upper()
        olho_valor = input("Cor dos olhos (A - Azuis, C - Castanhos): ").upper()
        cabelo_valor = input("Cor dos cabelos (L - Louros, P - Pretos): ").upper()

        # Armazenando os dados nas listas globais
        idade.append(idade_valor)
        genero.append(genero_valor)
        olho.append(olho_valor)
        cabelo.append(cabelo_valor)

def media_idades_olhos_castanhos_cabelos_pretos():
    soma_idades = 0
    contador = 0

    # Calcular a média das idades das pessoas com olhos castanhos e cabelos pretos
    for i in range(len(idade)):
        if olho[i] == 'C' and cabelo[i] == 'P':
            soma_idades += idade[i]
            contador += 1

    if contador > 0:
        return soma_idades / contador
    else:
        return 0  # Se não houver ninguém com olhos castanhos e cabelos pretos

def maior_idade():
    return max(idade)  # Retorna a maior idade na lista de idades

def contar_femininas_olhos_azuis_cabelos_louros():
    contador = 0
    # Contar as mulheres entre 18 e 35 anos com olhos azuis e cabelos louros
    for i in range(len(idade)):
        if genero[i] == 'F' and 18 <= idade[i] <= 35 and olho[i] == 'A' and cabelo[i] == 'L':
            contador += 1
    return contador

def main():
    cadastrar()

    # Calcula a média de idades das pessoas com olhos castanhos e cabelos pretos
    media = media_idades_olhos_castanhos_cabelos_pretos()
    print(f"Média de idades de pessoas com olhos castanhos e cabelos pretos: {media:.2f}")

    # Obtém a maior idade entre os habitantes
    maior = maior_idade()
    print(f"A maior idade entre os habitantes é: {maior}")

    # Conta a quantidade de mulheres com olhos azuis, cabelos louros, e idade entre 18 e 35 anos
    quantidade = contar_femininas_olhos_azuis_cabelos_louros()
    print(f"Quantidade de mulheres entre 18 e 35 anos, com olhos azuis e cabelos louros: {quantidade}")

# Chama a função principal
main()
```
